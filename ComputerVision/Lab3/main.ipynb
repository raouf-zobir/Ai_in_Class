# Lab 3: Corner and Edge Detection with Filter Application

#### Task 1: Image Pre-processing
#### Steps:
- Load an image and convert it to grayscale.
- Apply a Gaussian blur to reduce noise and improve filter performance.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('image.jpg')

# Display the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

# Show images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(gray, cmap='gray'), plt.title('Grayscale Image')
plt.subplot(1, 2, 2), plt.imshow(blurred, cmap='gray'), plt.title('Blurred Image')
plt.show()
```

#### Task 2: Edge Detection with Filters
#### Sobel Filter and Canny Edge Detection


```python
# Sobel Edge Detection
sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobelx, sobely)

# Canny Edge Detection
edges = cv2.Canny(blurred, 100, 200)

# Show images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Edge Detection')
plt.subplot(1, 2, 2), plt.imshow(edges, cmap='gray'), plt.title('Canny Edge Detection')
plt.show()
```

#### Task 3: Corner Detection with Harris Detector


```python
# Harris Corner Detection
harris_corners = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
harris_corners = cv2.dilate(harris_corners, None)

# Mark corners in red
corner_image = image.copy()
corner_image[harris_corners > 0.01 * harris_corners.max()] = [0, 0, 255]

# Show the image
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(corner_image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection')
plt.show()
```

#### Task 4: Analysis and Comparison
- Sobel filter detects edges based on gradients and is sensitive to noise.
- Canny filter applies edge thinning and is better at detecting strong edges.
- Harris corner detector finds corners based on intensity variation.
